```python
import pandas as pd
import numpy as np
import librosa
import matplotlib.pyplot as plt
import librosa.display
import IPython.display as ipd
```

```python
audio_file = "/content/OAF_back_fear.wav"
```

```python
ipd.Audio(audio_file)
```

```python
signal, sr = librosa.load(audio_file)
```

sr: sampling rate

```python
signal.shape
```

```python
mfccs = librosa.feature.mfcc(y=signal, sr=sr, n_mfcc=40)
```

```python
mfccs.shape
```

here 40 represents number of coefficents and 75 are number of frames

```python
plt.figure(figsize=(10,5))
librosa.display.specshow(mfccs,sr=sr,x_axis='time')
plt.colorbar(format='%+2.0f dB')
plt.show()
```

```python
mfcc_mean = np.mean(mfccs, axis=1)
```

```python
mfcc_mean.shape
```

For each audio I extract their Mel frequency cepstral coefficients and take mean over time frames of audio, this would be used as feature.
